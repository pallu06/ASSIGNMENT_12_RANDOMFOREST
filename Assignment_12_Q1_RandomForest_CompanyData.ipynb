# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.ensemble import IsolationForest
```

```python
# Importing the Datasets
comp_data = pd.read_csv("Company_Data.csv")
```

```python
comp_data
```

# OneHotEncoder 1st method 

```python
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(comp_data[['ShelveLoc','Urban','US']]).toarray())
```

```python
enc_df
```

```python
# merge with main df 
comp_final = comp_data.iloc[:,0:4].join(enc_df)
comp_final
```

# OneHotEncoder 2nd method using get_dummies

```python
comp_data1 = pd.get_dummies(comp_data,prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False
                         , drop_first=False, dtype= None) 
```

```python
comp_data1
```

```python
# converting numerical values to caterical
# classify sales below 10 and less than 10
comp_data1['Sales'] = np.where(comp_data1.Sales > 10 , 1 , 0)
```

```python
comp_data1
```

# Isolation Forest

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(comp_data1)
```

```python
# predictions
y_pred_outliers = clf.predict(comp_data1)
```

```python
#-1 for outliers and 1 for inliers.
y_pred_outliers
```
